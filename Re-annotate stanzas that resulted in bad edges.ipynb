```python
import string

import nltk
import pandas as pd
import pickle
```

```python
with open('bad_edges.pickle','rb') as f:
    bad_edges = pickle.load(f)
    
df = pd.read_csv("tsvs/tita_rhymes_poems.tsv", sep="\t")
df
```

```python
stanzas = [stanza.split("\n") for stanza in df.stanza]
```

```python
punctuation = string.punctuation + "...«»—"

def tokenize_and_line_ending_word(line): 
    tokens = [t for t in nltk.tokenize.word_tokenize(line, language='norwegian', preserve_line=False) if t not in punctuation]
    return tokens[-1]
```

```python
line_ending_words = [[tokenize_and_line_ending_word(line) for line in stanza] for stanza in stanzas]
```

```python
potentially_bad_stanza_i = {}

for w1, w2 in bad_edges:
    for i, lew in enumerate(line_ending_words):
        if w1 in lew and w2 in lew:
            potentially_bad_stanza_i[i] = (w1, w2)
        
```

```python
len(potentially_bad_stanza_i), len(bad_edges)
```

```python
potentially_bad = df.loc[potentially_bad_stanza_i.keys()]
potentially_bad
```

```python
s = "\n\n".join(potentially_bad.stanza)
with open("re_annotate/re_annotate.txt", "w+") as f:
    f.write(s)
```
